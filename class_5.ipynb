```python
day={'Saturday':1,'Sunday':2,'Monday':3,'Tuesday':4,'Wednesday':5,'Thursday':6,'Friday':7}
df['Day of the week']=df['Day of the week'].replace(day)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in df.columns:
  plt.figure(figsize=(8,5))
  plt.hist(df[col].dropna(),bins=10,color="red",edgecolor="black")
  plt.xlabel(col)
  plt.ylabel("frequency")
  plt.show()
```

```python
plt.pie(df['Traffic Situation'].value_counts(),labels=df['Traffic Situation'].value_counts().index,autopct='%1.1f%%')
plt.title('Traffic Situation distribution')
plt.show()
```

```python
group=df[['CarCount','BikeCount','BusCount','TruckCount']].sum()
plt.figure(figsize=(8,5))
plt.pie(group,labels=group.index,autopct='%1.1f%%')
plt.title('vichles type dirtibution')
plt.show()
```

```python
avg=df.groupby('Day of the week')['Total'].mean()
plt.figure(figsize=(8,5))
plt.bar(avg.index,avg.values,color='green')
plt.xlabel('Day of the week')
plt.ylabel('Average Traffic')
plt.title('Average Traffic by Day of the Week')
plt.show()
```

```python
df
```

```python
vichles_count=df.groupby('Day of the week')[['CarCount','BikeCount','BusCount','TruckCount']].sum()
plt.figure(figsize=(8,5))
vichles_count.plot(kind='bar')
plt.title('vichles count')
plt.xlabel('Day of the week')
plt.ylabel('count')

```

```python
colors=sns.color_palette('pastel')
plt.figure(figsize=(8,5))
plt.pie(df['Traffic Situation'].value_counts(),labels=df['Traffic Situation'].value_counts().index,autopct='%1.1f%%',colors=colors)
plt.title('Traffic Situation distribution')
plt.show()
```

```python
columns=['CarCount','BikeCount','BusCount','TruckCount']
sns.boxplot(data=df[columns])
plt.title('boxplot')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Traffic Situation',y='CarCount',data=df)
plt.xlabel('Traffic Situation')
plt.ylabel('CarCount')
plt.show()
```

```python
sns.lineplot(data=df,x='Time',y='CarCount')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, fmt='.2f')
plt.show()
```
